```python
import statistics
import numpy as np

numbers = [10, 20, 30, 40, 50, 20]

mean_value = statistics.mean(numbers)
median_value = statistics.median(numbers)
mode_value = statistics.mode(numbers)
variance_value = statistics.variance(numbers)
std_deviation = statistics.stdev(numbers)

print(mean_value)
print(median_value)
print(mode_value)
print(variance_value)
print(std_deviation)
```

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5, 6])

print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))

reshaped = arr.reshape(2, 3)
print(reshaped)

multiplied = arr * 2
print(multiplied)
```

```python
import numpy as np

A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print(A + B)
print(A - B)
print(np.dot(A, B))
print(A.T)
```

```python
from scipy import stats
import numpy as np

data = [10, 12, 9, 11, 10, 13, 12]

mean_value = stats.tmean(data)
print(mean_value)

t_stat, p_value = stats.ttest_1samp(data, 10)
print(t_stat)
print(p_value)
```

```python
from scipy import optimize
import numpy as np

f = lambda x: x**2 + 4*x + 4

result = optimize.minimize(f, 0)

print(result.x)
print(result.fun)
```
